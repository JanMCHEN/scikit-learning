```python
from sklearn import datasets
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt

from linear_regression.logistic_reg import Logistic

%matplotlib inline
```

```python
cancer = datasets.load_breast_cancer()

# 数据分割
x_train, x_test, y_train, y_test = train_test_split(cancer.data, cancer.target, test_size=0.2)

# 数据预处理
mean, max_, min_ = x_train.mean(axis=0), x_train.max(axis=0), x_train.min(axis=0) 
x_train = (x_train - mean) / (max_-min_+1e-10)

logisic = Logistic()
logisic.fit(x_train, y_train, iters=100, r=0.1, alpha=0.5)
plt.plot(range(logisic.args['iter count']), logisic.args['loss'])
plt.xlabel('iter count'), plt.ylabel('loss')
print('train score:', logisic.args['score'], '\ttest score:', logisic.score((x_test-mean)/(max_-min_+1e-10), y_test))
```
